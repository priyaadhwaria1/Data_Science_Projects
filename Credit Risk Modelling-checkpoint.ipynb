# ML Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier  
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
lr= LogisticRegression()
lr.fit(x_train, y_train)
predlr= lr.predict_proba(x_test)
predlr
```

```python
pred_lr= lr.predict(x_test)
print('Accuracy Score :', accuracy_score(y_test,pred_lr))
print('Confusion Matrix :', confusion_matrix(y_test,pred_lr))
print('Classification Report :', classification_report(y_test,pred_lr))
```

```python
dtc= DecisionTreeClassifier()
dtc.fit(x_train, y_train)
preddtc= dtc.predict(x_test)
print('Accuracy Score :', accuracy_score(y_test,preddtc))
print('Confusion Matrix :', confusion_matrix(y_test,preddtc))
print('Classification Report :', classification_report(y_test,preddtc))
```

```python
rfc= RandomForestClassifier()
rfc.fit(x_train, y_train)
predrfc= rfc.predict(x_test)
print('Accuracy Score :', accuracy_score(y_test,predrfc))
print('Confusion Matrix :', confusion_matrix(y_test,predrfc))
print('Classification Report :', classification_report(y_test,predrfc))
```

```python
xgb= XGBClassifier()
xgb.fit(x_train, y_train)
predxgb= xgb.predict(x_test)
print('Accuracy Score :', accuracy_score(y_test,predxgb))
print('Confusion Matrix :', confusion_matrix(y_test,predxgb))
print('Classification Report :', classification_report(y_test,predxgb))
```

```python
from sklearn.model_selection import cross_val_score

cvs3= cross_val_score(xgb, x_train, y_train, cv=7)
cvs4= cross_val_score(xgb, x_test, y_test, cv=7)

print('Cross Validation Score for train dataset is:', cvs3.mean())
print('Cross Validation Score for test dataset is:', cvs4.mean())
```

# Scorecard Development

```python
from optbinning import BinningProcess
from optbinning import Scorecard
from optbinning.scorecard import plot_auc_roc, plot_ks
```

```python
var= list(x_train.columns)
var
```

```python
bins= BinningProcess(var, selection_criteria= {"iv": {"min": 0.02, "max": 1}})
```

```python
scorecard= Scorecard(binning_process= bins, estimator= lr, scaling_method="pdo_odds", scaling_method_params={"pdo": 20, "odds": 30, "scorecard_points": 600})
```

```python
scorecard.fit(X= x_train, y= y_train.cb_person_default_on_file.values)
```

```python
scorecard1= scorecard.score(x_test)
scorecard1
```

```python
scorecard1.min(), scorecard1.max()
```

```python
df_pred_y_test= pd.DataFrame(scorecard.predict(x_test), columns=['Predict_y_test'])
df_pred_y_test.head()
```

```python
df_y_test= pd.DataFrame(y_test['cb_person_default_on_file'].values, columns=['Actual_y_test'])
df_y_test.head()
```

```python
df_prob_test= pd.DataFrame(scorecard.predict_proba(x_test), columns= ['Prob_Good_test','Prob_Bad_test'])
df_prob_test.head()
```

```python
df_score_test= pd.DataFrame(scorecard.score(x_test), columns=['Score_test'])
df_score_test.shape
```

```python
df_final= pd.concat([df_y_test, df_pred_y_test, df_prob_test, df_score_test], axis=1)
df_final.head()
```

```python
df_final['Score_test']= round(df_final['Score_test'])
df_final.head()
```

```python
print('Max_Score=',df_final['Score_test'].max(), 'Min_Score=',df_final['Score_test'].min())
```
