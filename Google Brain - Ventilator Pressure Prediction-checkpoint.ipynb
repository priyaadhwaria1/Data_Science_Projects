```python
#Imporving skewness of required columns
columns=['C','u_in','u_out','pressure']
from sklearn.preprocessing import power_transform
pt=power_transform(train_df[columns],method='yeo-johnson')
train_df[columns]=pd.DataFrame(pt,columns=train_df[columns].columns)
train_df.skew()
```

```python
#X-Y split
x=train_df.drop('pressure',axis=1)
y=train_df['pressure']
```

```python
#train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
#Scaling x values
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.transform(x_test)
```

```python
#Importing Libraries required for prediction
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
#Linear Regression
lr=LinearRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
print(r2_score(y_test,predlr))
```

```python
#DecisionTree Regression
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
preddtr=dtr.predict(x_test)
print(r2_score(y_test,preddtr))
```

```python
#Parameters used for DecisionTree Regression
dtr.get_params()
```

# Cross Validation Score

```python
#Cross validation for Linear Regression
from sklearn.model_selection import cross_val_score 
src1=cross_val_score(lr,x_train,y_train,cv=3)
print('Cross Validation Score:', src1.mean())
```

```python
#Cross validation for Decision Tree Regression
src3=cross_val_score(dtr,x_train,y_train,cv=3)
print('Cross Validation Score:', src3.mean())
```

# Hyper Tuning Paramete

```python
#Grid SearchCV for Decision Tree Regressor
from sklearn.model_selection import GridSearchCV
parameter={'ccp_alpha':[0],
 'criterion': ['mse'],
 'max_depth': [30],
 'max_features': [5],
 'max_leaf_nodes': [None],
 'min_impurity_decrease':[0.0],
 'min_impurity_split': [None],
 'min_samples_leaf': [10],
 'min_samples_split': [20],
 'min_weight_fraction_leaf': [0.0],
 'random_state': [200],
 'splitter': ['best']}
gcv=GridSearchCV(dtr,parameter,cv=3)
gcv.fit(x_train,y_train)
predgcv=gcv.best_estimator_.predict(x_test)
r2_score(y_test,predgcv)
```
