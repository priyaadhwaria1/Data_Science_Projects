# Train-Test Splitting

```python
#Split of train and test dataset
from sklearn.model_selection import train_test_split
covid_train, covid_test  = train_test_split(covid, test_size=0.2, random_state=42)
print('Train set is of size', covid_train.shape,'\nTest set is of size', covid_test.shape)
```

# Feature Engineering

```python
#Checking train dataset
covid_train.head()
```

```python
#Checking shape of train dataset
covid_train.shape
```

```python
#reset Index
covid_train=covid_train.reset_index(drop=True)
```

```python
#Converting Date column to ordinal type or int datatype
import datetime as dt
covid_train['date']=covid_train['date'].map(dt.datetime.toordinal)
```

```python
#Checking Datatype
covid_train.dtypes
```

```python
#Checking Null Values
covid_train.isnull().sum()
```

```python
#Removing columns with null values greater than 40% of total rows 
for features in covid_train.columns:
    if covid_train[features].isnull().sum()>(0.4*(len(covid_train))):
        covid_train.drop([features],axis=1,inplace=True)
```

```python
#Checking Null Values
covid_train.isnull().sum()
```

```python
#Dropping unnecessary columns
covid_train= covid_train.drop(['iso_code','continent','new_deaths_smoothed','new_deaths','new_deaths_smoothed_per_million','new_deaths_per_million','total_deaths_per_million','new_cases_smoothed','new_cases','new_cases_smoothed_per_million','new_cases_per_million','total_cases_per_million','new_tests_per_thousand','new_tests_smoothed','new_tests_smoothed_per_thousand','total_tests','total_tests_per_thousand'],axis=1)
```

```python
#Checking rest of null values
covid_train.isnull().sum()
```

```python
#Checking value counts of the column
covid_train['tests_units'].value_counts()
```

```python
#Replacings null values with mode values in all columns with datatype object
for features in covid_train.columns:
     if covid_train[features].isnull().sum()>=1 and covid_train[features].dtypes=='O':
            covid_train[features]=covid_train[features].fillna(covid_train[features].mode()[0])
     
```

```python
#Checking value counts of the column
covid_train['tests_units'].value_counts()
```

```python
#Checking rest of null values
covid_train.isna().sum()
```

```python
#Replacings null values with median value in all columns with datatype int or float
for features in covid_train.columns:
    if covid_train[features].isnull().sum()>=1 and covid_train[features].dtypes!='O':
        covid_train[features]=covid_train[features].fillna(covid_train[features].median())
```

```python
#Appling One Hot Encoding on the columns
covid_train=pd.get_dummies(covid_train,columns=['location','tests_units'],drop_first=True)
```

```python
#Checking  improved dataFrame
covid_train.head()
```

```python
#Rest index
covid_train=covid_train.reset_index(drop=True)
```

```python
#Checking statistics values of columns of dataframe
covid_train.describe()
```

# Correlation

```python
#Checking correlation of columns of dataframe
corr=covid_train.corr()
plt.figure(figsize=[22,12])
sns.heatmap(corr,annot=True)
```

```python
#Dropping unnecessary columns on the basis of correlation values
covid_train=covid_train.drop(['total_deaths','new_tests','aged_65_older','aged_70_older','median_age','population','extreme_poverty','female_smokers','hospital_beds_per_thousand','life_expectancy','tests_units_samples tested','location_India','human_development_index','location_Russia','location_United States','population_density'],axis=1)
```

Here it is clear that only 'total_cases' column has outliers.

```python
#Checking skewness of all columns of the dataset
covid_train.skew()
```

Here it is clear that only 'total_cases' column has skewness.

```python
#Imporving skewness of required columns
columns=['total_cases']

from sklearn.preprocessing import power_transform
pt=power_transform(covid_train[columns], method='yeo-johnson')
covid_train[columns]=pd.DataFrame(pt, columns=covid_train[columns].columns)
covid_train.skew()
```

```python
#Checking outliers of the columns with boxplot
plt.figure(figsize=[22,12])
covid_train.iloc[:,0:1].boxplot()
```

Power Transform will improve both skewness and outliers. Since 'total_cases' had outliers and skewness both, I applied Power Transform method on this column.

```python
#X-Y split
X=covid_train.drop(['total_cases'],axis=1)
Y=covid_train[['total_cases']]
```

```python
#train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=25,test_size=0.2)
```

```python
#Scaling x values
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.transform(x_test)
```

# ML Models

```python
#Importing Libraries required for prediction
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
#Linear Regression
lr=LinearRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
mse01 = mean_squared_error(y_test,predlr)
rmse01 = np.sqrt(mse01)
print(r2_score(y_test,predlr)*100)
print('rmse',rmse01)
```

```python
#Converting x_train value into quadratic form
poly=PolynomialFeatures(degree=2)
poly_x_train=poly.fit_transform(x_train)
poly_x_test=poly.transform(x_test)
```

```python
#Polynomial Regression
polylr=LinearRegression()
polylr.fit(poly_x_train,y_train)
predpoly=polylr.predict(poly_x_test)
mse02 = mean_squared_error(y_test,predpoly)
rmse02 = np.sqrt(mse02)
print(r2_score(y_test,predpoly)*100)
print('rmse',rmse02)
```

```python
#DecisionTree Regression
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
preddtr=dtr.predict(x_test)
mse03 = mean_squared_error(y_test,preddtr)
rmse03 = np.sqrt(mse03)
print(r2_score(y_test,preddtr)*100)
print('rmse',rmse03)
```

```python
##Parameters used for DecisionTree Regression
dtr.get_params()
```

```python
#RandomForest Regression
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
mse04 = mean_squared_error(y_test,predrfr)
rmse04 = np.sqrt(mse04)
print(r2_score(y_test,predrfr)*100)
print('rmse',rmse04)
```

```python
#Parameters used for RandomForest Regression
rfr.get_params()
```

```python
#Lasso Regression
lasso=Lasso(normalize=True)
lasso.fit(x_train,y_train)
predla=lasso.predict(x_test)
mse05 = mean_squared_error(y_test,predla)
rmse05 = np.sqrt(mse05)
print(r2_score(y_test,predla)*100)
print('rmse',rmse05)
```

```python
#Ridge Regression
ridge=Ridge(alpha=1.0,normalize=True)
ridge.fit(x_train,y_train)
predre=ridge.predict(x_test)
mse06 = mean_squared_error(y_test,predre)
rmse06 = np.sqrt(mse06)
print(r2_score(y_test,predre)*100)
print('rmse',rmse06)
```

```python
#Gradient Boosting Regression
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
predgbr=gbr.predict(x_test)
mse07 = mean_squared_error(y_test,predgbr)
rmse07 = np.sqrt(mse07)
print(r2_score(y_test,predgbr)*100)
print('rmse',rmse07)
```

```python
#XGB Regression
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
predxgb=xgb.predict(x_test)
mse08 = mean_squared_error(y_test,predxgb)
rmse08 = np.sqrt(mse08)
print(r2_score(y_test,predxgb)*100)
print('rmse',rmse08)
```

```python
#Parameters used for XGB Regression
xgb.get_params()
```

Since Decision Tree Regression, Random Forest Regression, Gradient Boosting Regression and XGB Regression algorithms has highest R2 score, so I will proceed with these 4 ML algorithms.

```python
#Cross validation for XGB Regression
from sklearn.model_selection import cross_val_score
scores3 = cross_val_score(xgb,x_train,y_train, scoring="neg_mean_squared_error", cv=10)
rmse_scores3 = np.sqrt(-scores3)
print('rmse mean:',rmse_scores3.mean())
print('rmse Standard Deviation:',rmse_scores3.std())
```

```python
#Cross validation for Gradient Boosting Regression
from sklearn.model_selection import cross_val_score
scores4 = cross_val_score(gbr,x_train,y_train, scoring="neg_mean_squared_error", cv=10)
rmse_scores4 = np.sqrt(-scores4)
print('rmse mean:',rmse_scores4.mean())
print('rmse Standard Deviation:',rmse_scores4.std())
```

Since, Cross validation rmse mean score of Decision Tree Regression, Random Forest Regression and XGB Regression is least, so I will proceed with these three ML Algorithms.

# Hyper Parameter Tuning

```python
#GridSearchCV for Decision Tree Regression
from sklearn.model_selection import GridSearchCV
parameter={'ccp_alpha': [0.004],
 'criterion': ['mse'],
 'max_depth': [20],
 'max_features': [2],
 'max_leaf_nodes': [None],
 'min_impurity_decrease': [0.0],
 'min_impurity_split': [None],
 'min_samples_leaf': [3],
 'min_samples_split': [3],
 'min_weight_fraction_leaf': [0.0],
 'random_state': [100],
 'splitter': ['best']}
gcv_dtr=GridSearchCV(dtr,parameter,cv=10)
gcv_dtr.fit(x_train,y_train)
predgcv_dtr=gcv_dtr.predict(x_test)
mse10 = mean_squared_error(y_test,predgcv_dtr)
rmse10 = np.sqrt(mse10)
print(r2_score(y_test,predgcv_dtr)*100)
print('rmse',rmse10)
```

```python
# Cross validation for tuned Decision Tree Regression
from sklearn.model_selection import cross_val_score
scores10 = cross_val_score(gcv_dtr,x_train,y_train, cv=10)
scores10.mean()
```

```python

#Grid SearchCV for Random Forest Regression
from sklearn.model_selection import GridSearchCV
parameter={'bootstrap': [False],
 'ccp_alpha': [0.004],
 'criterion': ['mse'],
 'max_depth': [10],
 'max_features': ['auto','log2','sqrt'],
 'max_leaf_nodes': [None],
 'max_samples': [20],
 'min_impurity_decrease': [0.0],
 'min_impurity_split': [None],
 'min_samples_leaf': [1],
 'min_samples_split': [2],
 'min_weight_fraction_leaf': [0.0],
 'n_estimators': [20],
 'n_jobs': [50],
 'oob_score': [False],
 'random_state': [200],
 'verbose': [0],
 'warm_start': [False]}
gcv_rfr=GridSearchCV(rfr,parameter,cv=10)
gcv_rfr.fit(x_train,y_train)
predgcv_rfr=gcv_rfr.predict(x_test)
mse20 = mean_squared_error(y_test,predgcv_rfr)
rmse20 = np.sqrt(mse20)
print(r2_score(y_test,predgcv_rfr)*100)
print('rmse',rmse20)
```

```python
#Cross validation for tuned Random Forest Regression
from sklearn.model_selection import cross_val_score
scores20 = cross_val_score(gcv_rfr,x_train,y_train, cv=10)
scores20.mean()
```

```python
#Grid SearchCV for XGB Regression
parameter={'objective': ['reg:squarederror'],
 'base_score': [0.05],
 'booster': ['gbtree'],
 'colsample_bylevel': [1],
 'colsample_bynode': [1],
 'colsample_bytree': [1],
 'gamma': [1],
 'gpu_id': [1],
 'importance_type': ['gain'],
 'learning_rate': [0.09],
 'max_delta_step': [0],
 'max_depth': [5],
 'min_child_weight': [1],
 'n_estimators': [100],
 'n_jobs': [10],
 'num_parallel_tree': [4],
 'random_state': [100],
 'reg_alpha': [0],
 'reg_lambda': [2],
 'scale_pos_weight': [1],
 'subsample': [1],
 'tree_method': ['exact'],
 'validate_parameters': [3],
 'verbosity': [None]}
gcv_xgb=GridSearchCV(xgb,parameter,cv=10)
gcv_xgb.fit(x_train,y_train)
predgcv_xgb=gcv_xgb.predict(x_test)
mse30 = mean_squared_error(y_test,predgcv_xgb)
rmse30 = np.sqrt(mse30)
print(r2_score(y_test,predgcv_xgb)*100)
print('rmse',rmse30)
```
